```python
import google.generativeai as genai
from serpapi import GoogleSearch
from youtube_transcript_api import YouTubeTranscriptApi
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
import spacy
import dateparser
```

```python
# Gemini API key
gemini_api_key = ""

# SerpAPI key
serpapi_key = ""
```

```python
def get_transcript(video_url):
    video_id = video_url.split("v=")[-1]
    transcript = YouTubeTranscriptApi.get_transcript(video_id)
    text = " ".join([item['text'] for item in transcript])
    return text

# Example
video_url = "https://www.youtube.com/watch?v=JIbIYCM48to&t=99s"
transcript_text = get_transcript(video_url)
print(transcript_text[:1000])
```

```python
# Load spaCy and Hugging Face NER model
nlp = spacy.load("en_core_web_sm")
model_name = "dslim/bert-base-NER"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

def extract_factual_claims(transcript_text):
    doc = nlp(transcript_text)
    extracted_claims = []
    for sent in doc.sents:
        sentence_text = sent.text.strip()
        if len(sentence_text) < 5:
            continue
        entities = ner_pipeline(sentence_text)
        if entities:
            claim_info = {
                "sentence": sentence_text,
                "entities": []
            }
            for ent in entities:
                entity_data = {
                    "text": ent['word'],
                    "type": ent['entity_group'],
                    "score": ent['score']
                }
                if ent['entity_group'] == "DATE":
                    parsed_date = dateparser.parse(ent['word'])
                    entity_data["parsed_date"] = str(parsed_date) if parsed_date else None
                claim_info["entities"].append(entity_data)
            extracted_claims.append(claim_info)
    return extracted_claims

# Example
claims = extract_factual_claims(transcript_text)
for c in claims[:5]:
    print(c)
```

# Using Gimini to Generate Link:

```python
import google.generativeai as genai

def fact_check_with_gemini(claim, api_key):
    try:
        # Configure the API
        genai.configure(api_key=api_key)

        # Initialize the model (use a valid model name)
        model = genai.GenerativeModel("models/gemini-1.5-flash")  # or "gemini-pro"

        # Structured prompt for better responses
        prompt = f"""
        **Fact-Checking Task:**
        Claim: "{claim}"

        **Instructions:**
        1. Determine if the claim is **True**, **False**, or **Unverifiable**.
        2. Provide a **1-2 sentence explanation**.
        3. If possible, include three **credible source URL** (official sites).

        **Respond in this format:**
        - Verdict: [True/False/Unverifiable]
        - Explanation: [Brief reasoning]
        - Source: [URL or "Not available"]
        """

        # Generate the response
        response = model.generate_content(prompt)
        return response.text.strip()

    except Exception as e:
        return f"Error: {str(e)}"

# Example usage

result = fact_check_with_gemini("The Eiffel Tower is in Rome.", gemini_api_key)
print(result)
```

# Using Duck Duck (Faks de):

```python
from duckduckgo_search import DDGS
import time
import random

def fact_check_with_Duck_Duck(claim, max_results=3):
    """
    Comprehensive fact-checking that searches:
    - First checks fact-checking sites
    - Then falls back to general web search
    - Includes rate limiting
    """
    result = {
        'claim': claim,
        'verdict': 'Unverifiable',
        'explanation': 'Initializing search...',
        'sources': []
    }

    try:
        with DDGS() as ddgs:
            # Try specialized fact-check sites first
            fact_check_sites = [
            ]

            # Search each fact-check site with delay
            for site in fact_check_sites:
                try:
                    query = f"{site} {claim}"
                    results = list(ddgs.text(query, max_results=1))
                    time.sleep(random.uniform(0, 1))  # Random delay

                    if results:
                        result['sources'].extend([{
                            'title': r.get('title'),
                            'url': r.get('href'),
                            'snippet': r.get('body','')[:200] + '...',
                            'type': 'fact-check'
                        } for r in results])

                        if len(result['sources']) >= max_results:
                            break
                except:
                    continue

            # If not enough fact-check results, supplement with general web
            if len(result['sources']) < max_results:
                general_query = f"fact check {claim}"
                general_results = list(ddgs.text(general_query,
                                           max_results=max_results - len(result['sources'])))
                time.sleep(1)

                result['sources'].extend([{
                    'title': r.get('title'),
                    'url': r.get('href'),
                    'snippet': r.get('body','')[:200] + '...',
                    'type': 'general-web'
                } for r in general_results])

            # Analyze results for verdict
            if result['sources']:
                content = ' '.join([s['title'] + s['snippet'] for s in result['sources']]).lower()

                if any(word in content for word in ['false', 'incorrect', 'wrong', 'debunked']):
                    result.update({
                        'verdict': 'False',
                        'explanation': 'Multiple sources dispute this claim'
                    })
                elif any(word in content for word in ['true', 'correct', 'accurate', 'confirmed']):
                    result.update({
                        'verdict': 'True',
                        'explanation': 'Credible sources support this claim'
                    })
                else:
                    result.update({
                        'verdict': 'Uncertain',
                        'explanation': 'Found mixed or inconclusive evidence'
                    })

    except Exception as e:
        result.update({
            'verdict': 'Error',
            'explanation': f"Search failed: {str(e)}"
        })

    return result

# Example usage
result = fact_check_with_Duck_Duck("The Eiffel Tower is in Rome.")
print(f"\nClaim: {result['claim']}")
print(f"Verdict: {result['verdict']}")
print(f"Explanation: {result['explanation']}")

if result['sources']:
    print("\nSources:")
    for i, source in enumerate(result['sources'], 1):
        print(f"{i}. [{source['type']}] {source['title']}")
        print(f"   {source['url']}")
        print(f"   {source['snippet']}")
```

# Using SerpAPI (Faks de):

```python
def fact_check_with_serpapi(claim, api_key, max_num = 3):
    search = GoogleSearch({
        "q": claim,
        "api_key": api_key,
        "num": max_num+1
    })
    results = search.get_dict()
    organic_results = results.get("organic_results", [])
    facts = []
    for result in organic_results:
        facts.append({
            "title": result.get("title"),
            "snippet": result.get("snippet"),
            "link": result.get("link")
        })
    return facts

# Example
fact_check_with_serpapi("The Eiffel Tower is in paris.", serpapi_key)
```

```python
def hybrid_fact_check(claim, serpapi_key, gemini_key):
    results = fact_check_with_gemini(claim, gemini_key)
    if results:
        return {"method": "Gimini LLM", "results": results}
    else:
        verdict = fact_check_with_gemini(claim, serpapi_key)
        return {"method": "SerpAPI", "verdict": verdict}

# Example
hybrid_fact_check("Barack Obama was born in Kenya.", serpapi_key, gemini_api_key)
```
